```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use("seaborn-darkgrid")
```

```python
data = pd.read_csv("logs_no_vae_ppo.txt", header=None)
data.head()
```

```python
data.iloc[:,0] -= 1
```

```python
data.iloc[:,1].rolling(5).mean().fillna(data.iloc[:,1])
n = len(data)+1
```

```python
plt.plot( range(1, n), data.iloc[:,1], label="Reward per Epoch" )
plt.plot( range(1, n), data.iloc[:,1].rolling(5).mean().fillna(data.iloc[:,1]), label="Avg Reward (5 Epochs)")
#plt.plot( range(1, n), data.iloc[:,1].expanding(5).mean(), label="Average Reward (5 Epochs)")

plt.xlabel("Epochs")
plt.ylabel("Reward")
plt.legend()
```

```python
plt.plot( range(1, n), data.iloc[:,2], label="Avg Episode Length per Epoch" )
plt.plot( range(1, n), data.iloc[:,2].rolling(5).mean().fillna(data.iloc[:,2]), label="Avg Episode Length (5 Epochs)")
plt.xlabel("Epochs")
plt.ylabel("Episode Length")
plt.legend()
```

```python
plt.plot( range(1, n), data.iloc[:,3], label="Episodes per Epoch" )
plt.plot( range(1, n), data.iloc[:,3].rolling(5).mean().fillna(data.iloc[:,3]), label="Episodes per Epoch (5 Epochs)")
plt.xlabel("Epochs")
plt.ylabel("Number of Episodes")
plt.legend()
```
